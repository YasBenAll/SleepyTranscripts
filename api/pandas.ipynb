```python
import pandas as pd
import json
df = pd.read_excel("youtube-link.xlsx", header=None)
```

```python
d = {}
for i in range(len(df)):
    d[df[0][i]] = df[1][i]
d
```

```python
with open("slug_keys.json", "r") as f:
    slugConverter = json.load(f)
for i in list(slugConverter.keys()):
    print(i, slugConverter[i])
```

```python
youtubeLinks = {}
for i,j in zip(slugConverter.keys(), d.values() ):
    youtubeLinks[i] = j

with open("youtubeLinks.json", "w") as f:
    json.dump(youtubeLinks, f, ensure_ascii=False)
```

```python
for i in slugConverter.keys():
    print(i)
```

```python
episodeData = {}

for i in slugConverter.keys():
    name = slugConverter[i]
    try:
        youtubeLink = youtubeLinks[i]
    except:
        youtubeLink = None
    episodeData[i] = {"name": name, "youtubeLink": youtubeLink}
episodeData

with open("episode_metadata.json", "w") as f:
    json.dump(episodeData, f, ensure_ascii=False)
```
